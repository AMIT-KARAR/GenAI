1. Write a set of codes in python to accept an input from user and find the Sum of squares of first n natural numbers.

```python
n = int(input("Enter a natural number n: ")) 
sum_of_squares = 0 
for i in range(1, n+1): 
    sum_of_squares += i**2 
    
print(f"Sum of squares of first {n} natural numbers is: {sum_of_squares}")
```

```python
num = int(input("Enter a natural number: "))

print(f"Prime numbers between 1 and {num} are:")

for i in range(2, num+1):   
    for j in range(2, i):
        if i % j == 0:
            break
    else:
        print(i, end=" ")
```

Write a program to find the largest & smallest of all the numbers in a list. Also, find the average number in a list.

```python
numbers = []

while True:
    num = input("Enter a number (hit --> 'q' to quit): ")
    if num.lower() == 'q':
        break
    try:  
        num=int(num)      
        numbers.append(num)
    except ValueError:
        print("Invalid input — please enter a number or 'q' to quit.")

if not numbers:
    print("No numbers were entered")
else:
    smallest = min(numbers)
    largest = max(numbers)
    average = sum(numbers) / len(numbers)
    print(f"You have given bellow {len(numbers)} numbers:\n {numbers}")
    print(f"Smallest: in --> {numbers}-->", smallest)
    print(f"Largest:  in --> {numbers}-->", largest)
    print(f"Average:  in --> {numbers}-->", average)
    
```

List And Map Example:

Write a program to find the square of each item in a list using map & lambda function. The list should take input from the users, probably 7 to 9 number in the list.

```python
numbers=input("Enter numbers by space: ")

num = list(map(int, numbers.split()))

squares = list(map(lambda x: x**2, num))

print("You Have Entered:", numbers)

print("You Given Numbers Squares:", squares)
```

Write a Python program to accept numbers from users as a and b determine if a is a prime factor of b and vice versa.

```python
def prime(n):                     #Define a function to check prime
    if n <= 1:                    #Number Should be Gater than 1 else return false as not prime
        return False
    for i in range(2, n):         #Iterate from 2 to the number ex if n=5 loop will iterate from 2 to 4 as if for prime 5 should be div by 1 and 5.
        if n % i == 0:            
            return False          #If division happen return not prime
    return True                   #If num gater than 0 and only div with 1 and it self return Prime


a = int(input("Enter number a: "))          #Taking two num from user
b = int(input("Enter number b: "))


if b % a == 0 and prime(a):                 # check div for prime factor and factor also a prime number
    print(a, "is a prime factor of", b)
else:
    print(a, "is NOT a prime factor of", b)

if a % b == 0 and prime(b):
    print(b, "is a prime factor of", a)
else:
    print(b, "is NOT a prime factor of", a)
```
