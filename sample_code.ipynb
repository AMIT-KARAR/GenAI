1. Write a set of codes in python to accept an input from user and find the Sum of squares of first n natural numbers.

```python
n = int(input("Enter a natural number n: ")) 
sum_of_squares = 0 
for i in range(1, n+1): 
    sum_of_squares += i**2 
    
print(f"Sum of squares of first {n} natural numbers is: {sum_of_squares}")
```

2. Write the set of python codes to accept a natural number and display the factorial of the number.

```python
n = int(input("Enter a natural number n: ")) 

fact_of_n=1
for i in range(1, n+1):
    fact_of_n *= i 

print(f"factorial of the number is {fact_of_n}")
```

3. Write a set of codes in python to accept a natural number "num" and display all the prime numbers between 1 to "num".

```python
num = int(input("Enter a natural number: "))

print(f"Prime numbers between 1 and {num} are:")

for i in range(2, num+1):   
    for j in range(2, i):
        if i % j == 0:
            break
    else:
        print(i, end=" ")
```

Decimal to binary converter

```python
num = int(input("Enter a positive decimal Number: "))

if num == 0:
    print("You have enter 0 please try with a positive decimal number!!!")
else:
    bits = []
    while num > 0:
        bits.append(str(num % 2))
        num //= 2
    bits.reverse()
    print("".join(bits))
    
    

```

Write a program to find the largest & smallest of all the numbers in a list. Also, find the average number in a list.

```python
numbers = []

while True:
    num = input("Enter a number (hit --> 'q' to quit): ")
    if num.lower() == 'q':
        break
    try:  
        num=int(num)      
        numbers.append(num)
    except ValueError:
        print("Invalid input — please enter a number or 'q' to quit.")

if not numbers:
    print("No numbers were entered")
else:
    smallest = min(numbers)
    largest = max(numbers)
    average = sum(numbers) / len(numbers)
    print(f"You have given bellow {len(numbers)} numbers:\n {numbers}")
    print(f"Smallest: in --> {numbers}-->", smallest)
    print(f"Largest:  in --> {numbers}-->", largest)
    print(f"Average:  in --> {numbers}-->", average)
    
```
