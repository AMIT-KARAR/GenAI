```python
from nltk.stem import WordNetLemmatizer
lematization=WordNetLemmatizer()
import nltk
nltk.download('wordnet')
```

```python
'''
POS- Noun-n
verb-v
adjective-a
adverb-r
'''

lematization.lemmatize("looking", pos="n")
```

```python
lematization.lemmatize("looking", pos="v")
```

```python
words=words=["eating","eats","eaten","writing","writes","programming","programs","history","finally","finalized"]

print(words)
```

```python
for word in words:
  print(word+ "----------------->"+lematization.lemmatize(word, pos= "v"))
```
